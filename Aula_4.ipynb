Eliminação de Gauss Ingênua

```python
import numpy as np

A = np.array([[2, 4,-6], [4, 2, 2], [2, 8,-4]])
b = np.array([[10], [16], [24]])
Aum = np.hstack((A, b))
n = len(b)

# Eliminação Progressiva
for i in range(n-1):
    for j in range(i+1,n):
        fator = Aum[j,i]/Aum[i,i];
        Aum[j,i:n+1] = Aum[j,i:n+1] - fator*Aum[i,i:n+1]

# Substituição Regressiva
x = np.zeros(n)

x[n-1] = Aum[n-1,n]/Aum[n-1,n-1]

for i in range(n-2,-1,-1):
    soma = 0
    for j in range(i+1,n):
        soma += Aum[i,j] * x[j]
    x[i] = (Aum[i,n] - soma) / Aum[i,i]
print(x)
```

Gauss-Seidel

Método de Jacobi

```python
import numpy as np
import math

A = np.array([[3, -0.1, -0.2], [0.1, 7, -0.3], [0.3, -0.2, 10]])
b = np.array([[7.85], [-19.3], [71.4]])
k = len(b)
x_old = np.ones(k)
x_new = np.zeros(k)

n = 6
Eppara = 0.5*10**(2-n)

Epest = np.linspace(100, 100, k)

while max(Epest) >= Eppara:
  for i in range(k):
    soma = 0
    for j in range(k-1):
      if i != j:
        soma += A[i,j] * x_old[j]
      else:
        soma = 0
    x_new[i] = (b[i] - soma) / A[i,i]
  Epest = abs((x_new - x_old)/x_new)*100
  x_old = x_new
```

Newton-Raphson

```python
import numpy as np

def f1(x1, x2):
  return x1**2 - x1*x2 - 10

def f2(x1, x2):
  return x2 + 3*x1*(x2**2) - 57

def df1x1(x1, x2):
  return 2*x1 - x2

def df1x2(x1, x2):
  return -x1

def df2x1(x1, x2):
  return 3*x2**2

def df2x2(x1, x2):
  return 1 + 6*x1*x2

x_old = np.array([1.5, 3.5])  #[x1old, x2old]
x_new = x_old

Epest = np.linspace(10, 10, 100)
n = 6
Eppara = 0.5*10**(2-n)


while max(Epest) >= Eppara:
  x1_new = x_old[0] - ((f1(x_old[0],x_old[1])*df2x2(x_old[0], x_old[1]) - f2(x_old[0], x_old[1])*df1x2(x_old[0], x_old[1]))/ (df1x1(x_old[0], x_old[1])*df2x2(x_old[0], x_old[1]) - df1x2(x_old[0], x_old[1])*df2x1(x_old[0], x_old[1])))
  x2_new = x_old[1] - ((f2(x_old[0],x_old[1])*df1x1(x_old[0], x_old[1]) - f1(x_old[0], x_old[1])*df2x1(x_old[0], x_old[1]))/ (df1x1(x_old[0], x_old[1])*df2x2(x_old[0], x_old[1]) - df1x2(x_old[0], x_old[1])*df2x1(x_old[0], x_old[1])))

  x_new = np.array([x1_new, x2_new])
  Epest = abs((x_new - x_old)/x_new)*100
  x_old = x_new
```
