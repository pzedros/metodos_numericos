Método Gráfico

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.sin(4*x) + np.cos(3*x)

# Definir intervalo de x
xmin, xmax = 0, 5
ns = 200
x = np.linspace(xmin, xmax, ns)
y = f(x)

# Plotar a função
plt.figure(figsize=(8,5))
plt.plot(x, y, label='$f(x) = \sin(4x) + \cos(3x)$')
plt.axhline(0, color='r', linewidth=1, linestyle='--')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.title("Método Gráfico")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

Método Busca Incremental

```python
import numpy as np
import matplotlib.pyplot as plt

def busca_incremental(func,xmin,xmax,ns):
  x = np.linspace(xmin,xmax,ns)
  f = func(x)
  xb = []

  for k in range(0,len(x) - 1):
    if f[k]*f[k+1] < 0:
      xb.append([x[k], x[k+1]])

  return np.array(xb)

# Exemplo
def exemplo_func(x):
  return np.sin(4*x) + np.cos(3*x)

xmin = 0
xmax = 5
ns = 200

subintervalos = busca_incremental(exemplo_func,xmin,xmax,ns)
print('Subintervalos:',subintervalos)

x = np.linspace(xmin,xmax,ns)
y = exemplo_func(x)

# Plotar a função
plt.figure(figsize=(8,5))
plt.plot(x, y, label='$f(x) = \sin(4x) + \cos(3x)$')
plt.axhline(0, color='r', linewidth=1, linestyle='--')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.title("Método Busca Incremental")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

Método Newton Raphson

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math

def f(x):
    return np.sin(4 * x) + np.cos(3 * x)

def df(x):
    return 4 * np.cos(4 * x) - 3 * np.sin(3 * x)

# Precisão desejada
n = 6
Eppara = 0.5 * (10 ** (2 - n))

# Subintervalos fornecidos
Subintervalos = np.array([
    [0.65326633, 0.67839196],
    [1.55778894, 1.58291457],
    [2.46231156, 2.48743719],
    [3.34170854, 3.36683417],
    [4.24623116, 4.27135678],
    [4.69849246, 4.72361809]])

# Armazenar raízes encontradas
raizes = []

# Loop sobre os subintervalos
for i, intervalo in enumerate(Subintervalos):
    a, b = intervalo
    x_old = (a + b) / 2  # chute inicial como ponto médio
    Ept = 100
    iteracoes = 0

    while Ept >= Eppara:
        try:
            x_new = x_old - (f(x_old) / df(x_old))
        except ZeroDivisionError:
            print(f"Derivada zero em x = {x_old}. Método falhou no intervalo {i+1}.")
            break
        Ept = abs((x_new - x_old) / x_new) * 100
        x_old = x_new
        iteracoes += 1

    raizes.append(x_new)
    print(f"Subintervalo {i+1}: Raiz ≈ {x_new:.8f}, Iterações = {iteracoes}")

# Plot do gráfico
x_vals = np.linspace(0, 2 * np.pi, 1000)
y_vals = f(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label=r'$f(x) = \sin(4x) + \cos(3x)$', color='blue')
plt.axhline(0, color='black', linestyle='--', linewidth=0.8)

# Marcar raízes
for i, raiz in enumerate(raizes):
    plt.plot(raiz, f(raiz), 'ro')
    plt.text(raiz, f(raiz) + 0.2, f'{i+1}', ha='center', fontsize=9)

plt.title('Gráfico de f(x) e raízes encontradas')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

Método da Secante Modificada

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def f(x):
    return np.sin(4 * x) + np.cos(3 * x)

# Precisão desejada
n = 6
Eppara = 0.5 * (10 ** (2 - n))
max_iter = 50  # Número máximo de iterações

# Subintervalos fornecidos
Subintervalos = np.array([
    [0.65326633, 0.67839196],
    [1.55778894, 1.58291457],
    [2.46231156, 2.48743719],
    [3.34170854, 3.36683417],
    [4.24623116, 4.27135678],
    [4.69849246, 4.72361809]
])

# Armazenar raízes encontradas
raizes = []
dados_iteracoes = []

# Loop sobre os subintervalos
for i, intervalo in enumerate(Subintervalos):
    a, b = intervalo
    x0, x1 = a, b
    # x_old_old = a  # primeiro chute
    # x_old = b      # segundo chute
    Ept = 100
    # iteracoes = 0
    iteracao = 0
    convergiu = False
    print(f"\nSubintervalo {i+1}: [{a:.8f}, {b:.8f}]")
    while Ept >= Eppara and iteracao < max_iter:
        # f_old = f(x_old)
        # f_old_old = f(x_old_old)

        f0, f1 = f(x0), f(x1)

        if abs(f1 - f0) < 1e-12:  # Evita divisão por zero
            print(f"Diferença entre f(x) muito pequena na iteração {iteracao}")
            break

        # if f_old - f_old_old == 0:
        #     print(f"Divisão por zero na iteração {iteracoes} do subintervalo {i+1}")
        #     break
        x_new = x1 - f1 * (x1 - x0) / (f1 - f0)
        Ept = abs((x_new - x1) / x_new) * 100

        print(f"Iteração {iteracao}: x = {x_new:.8f}, f(x) = {f(x_new):.2e}, Ept = {Ept:.2e}%")
        x0, x1 = x1, x_new
        # x_new = x_old - f_old * (x_old_old - x_old) / (f_old_old - f_old)
        # Ept = abs((x_new - x_old) / x_new) * 100
        # x_old_old = x_old
        # x_old = x_new
        iteracoes += 1

        if abs(f(x_new)) < Eppara: # Verifica se f(x) está próximo de zero
            convergiu = True
            break

    if convergiu or Ept < Eppara:
        raizes.append(x_new)
        dados_iteracoes.append(iteracao)
        print(f"Convergência alcançada: Raiz ≈ {x_new:.8f}, Iterações = {iteracao}")
    else:
        print(f"O método não convergiu após {iteracao} iterações")
    # print(f"Subintervalo {i+1}: Raiz ≈ {x_new:.8f}, Iterações = {iteracoes}")

#Plot do gráfico
x_vals = np.linspace(0, 2 * np.pi, 1000)
y_vals = f(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label=r'$f(x) = \sin(4x) + \cos(3x)$', color='blue')
plt.axhline(0, color='black', linestyle='--', linewidth=0.8)

# Marcar raízes
for i, raiz in enumerate(raizes):
    plt.plot(raiz, f(raiz), 'ro')
    plt.text(raiz, f(raiz) + 0.2, f'{i+1}', ha='center', fontsize=9)

plt.title('Gráfico de f(x) e raízes encontradas com o método da Secante')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```
