```python
import numpy as np

# Definindo a função
def f(x):
    return 0.2 + 25*x - 200*x**2 + 675*x**3 - 900*x**4 + 400*x**5

# Limites da integral
a = 0
b = 0.8
valor_exato = 1.640533

n = 1
h = (b - a) / n

x0 = a
xn = b

# Aplicando a Regra do Trapézio (simples)
I = (h / 2) * (f(x0) + f(xn))

# Erro percentual verdadeiro
Ept = abs((valor_exato - I) / valor_exato) * 100

print(f"Integral aproximada (Regra do Trapézio com n = {n}): {I:.2f}")
print(f"Erro percentual verdadeiro: {Ept:.2f}%")
```

```python
import numpy as np

# f(x)
def f(x):
    return 0.2 + 25*x - 200*x**2 + 675*x**3 - 900*x**4 + 400*x**5

# Limites de integração
a = 0
b = 0.8
I_exato = 1.640533

# Número de subintervalos (n = 2)
n = 2
h = (b - a) / n

# Pontos de avaliação
x0 = a
x1 = a + h
x2 = b

# Avaliação da função nos pontos
f0 = f(x0)
f1 = f(x1)
f2 = f(x2)

# Regra do Trapézio
I_trapezio = (h / 2) * (f0 + 2 * f1 + f2)

# Regra de Simpson 1/3
I_simpson = (h / 3) * (f0 + 4 * f1 + f2)


# Resultados
print(f"--- Resultados com n = {n} ---")
print(f"Integral (Trapézio): {I_trapezio:.6f}")
print(f"Integral (Simpson 1/3): {I_simpson:.6f}")
print(f"Valor exato: {I_exato}")

```

Regra 1/3 de simpson múltipla exemplo 4

```python
import numpy as np

# Definindo a função f(x)
def f(x):
    return 0.2 + 25*x - 200*x**2 + 675*x**3 - 900*x**4 + 400*x**5

# Limites de integração
a = 0
b = 0.8
I_exato = 1.640533

# Número de subintervalos (deve ser par)
n = 2

# Função da Regra de Simpson 1/3 Múltipla
def simpson13(f, a, b, n):
    h = (b - a) / n
    s0 = 0
    s1 = 0

    if n % 2 == 0:
        for i in range(1, n, 2):
            x = a + i * h
            s0 += f(x)

        for j in range(2, n-1, 2):
            xx = a + j * h
            s1 += f(xx)

        return (h / 3) * (f(a) + 4 * s0 + 2 * s1 + f(b))
    else:
        print("Erro: n deve ser par para aplicar a Regra de Simpson 1/3 Múltipla.")
        return None

# Cálculo da integral
I_simpson = simpson13(f, a, b, n)

# Cálculo do erro percentual
erro_percentual = abs((I_exato - I_simpson) / I_exato) * 100

# Resultados
print(f"Integral (Simpson 1/3 Múltipla): {I_simpson:.6f}")
print(f"Valor exato: {I_exato}")
print(f"Erro percentual: {erro_percentual:.4f}%")
```

# Exemplo 6: Fórmula de Gauss-Legendre de dois pontos

```python
import numpy as np

# Definindo a função
def f(x):
    return x**2 + 1

# Valor exato da integral
I_exato = 8/3

# Nós de Gauss-Legendre de 2 pontos
x1 = -1/np.sqrt(3)
x2 = 1/np.sqrt(3)

# Pesos (para n=2, ambos são 1)
I_gauss = f(x1) + f(x2)

# Erro percentual verdadeiro
erro_percentual_gauss = abs((I_exato - I_gauss) / I_exato) * 100

# Resultados
print(f"Integral (Gauss 2 pontos): {I_gauss:.4f}")
print(f"Valor exato: {I_exato}")
print(f"Erro percentual verdadeiro: {erro_percentual_gauss:.4f}%")
```
