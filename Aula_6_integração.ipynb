Regra do Trapézio

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return 0.2 + 25*x - 200*x**2 + 675*x**3 - 900*x**4 + 400*x**5

a = 0
b = 0.8
#valor verdadeiro
u = 1.640533


IT = (b-a)*(f(a)+f(b))/2
print(f'integral: {IT}')
Ept_IT = abs((u - IT)/u) * 100
print(f'erro: {Ept_IT}')

x = np.linspace(a,b,100)
plt.plot(x, f(x))
plt.grid()
```

Regra do Trapézio Múltipla

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return 0.2 + 25*x - 200*x**2 + 675*x**3 - 900*x**4 + 400*x**5

x0 = 0
xn = 0.8
n = 50 #numero de trapezios
h = (xn - x0)/n

#valor verdadeiro
u = 1.640533
soma = 0

for i in range(1, n):
  xi = x0 + h*i
  soma = soma + f(xi)

  ITM = h/2 * (f(x0) + 2*soma + f(xn))
  Ept_ITM = abs((u - ITM)/u) * 100

print(ITM)
print(Ept_ITM)

x = np.linspace(x0,xn,100)
plt.plot(x, f(x), '-r')
plt.grid()
```

Regra 1/3 de Simpson Múltipla

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return 0.2 + 25*x - 200*x**2 + 675*x**3 - 900*x**4 + 400*x**5

x0 = 0
xn = 0.8
n = 100 #numero de trapezios
h = (xn - x0)/n

#valor verdadeiro
u = 1.640533
somai = 0
somaj = 0

for i in range(1, n, 2):
  xi = x0 + h*i
  somai = somai + f(xi)

for j in range(2, n-1, 2):
  xj = x0 + h*j
  somaj = somaj + f(xj)

I = (xn - x0)*((f(x0) + 4*somai + 2*somaj + f(xn))/(3*n))
print(I)

x = np.linspace(x0,xn,100)
plt.plot(x, f(x), '-m')
plt.grid()
```
