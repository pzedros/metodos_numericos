```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 2, 4, 6, 9, 11, 12, 15, 17, 19])
y = np.array([5, 6, 7, 8, 9, 10, 10, 11, 12, 12])

n = len(x)
sum_xi = sum(x)
sum_xi_2 = sum(x**2)
sum_xi_yi = sum(x * y)
sum_yi = sum(y)

A = np.array([[n, sum_xi], [sum_xi, sum_xi_2]])
b = np.array([[sum_yi], [sum_xi_yi]])

ai = np.linalg.inv(A) @ b
print(ai)

plt.plot(x, y, 'or', label='dados experimentais')
plt.plot(x, ai[0] + ai[1] * x, "-b", label="$y = a_0 + a_1x$")
plt.xlabel("$x$")
plt.ylabel("$y(x)$")
plt.legend()
plt.grid()
plt.text(2, 10, r"$R^2 = \dfrac{S_T - S_R}{S_T}$", fontsize=10, bbox=dict(facecolor='white', edgecolor='red')) ##retirar depois
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 2, 4, 6, 9, 11, 12, 15, 17, 19])
y = np.array([5, 6, 7, 8, 9, 10, 10, 11, 12, 12])

n = len(x)
sum_xi = sum(x)
sum_xi_2 = sum(x**2)
sum_xi_yi = sum(x * y)
sum_yi = sum(y)

A = np.array([[n, sum_xi], [sum_xi, sum_xi_2]])
b = np.array([[sum_yi], [sum_xi_yi]])

ai = np.linalg.inv(A) @ b
print(ai)

plt.plot(x, y, 'or', label='dados experimentais')
plt.plot(x, ai[0] + ai[1] * x, "-b", label="$y = a_0 + a_1x$")
plt.xlabel("$x$")
plt.ylabel("$y(x)$")
plt.legend()
plt.grid()
plt.show()

#Calculo do R**2
y_medio=np.sum(y)/n
Sr=np.sum(y-ai[0]-ai[1]*x)
St=np.sum((y-y_medio)**2)
R_2=1-Sr/St
print(R_2)
```

Exercício 7 - Interpolação Quadrática de Newton.

```python
import numpy as np
import matplotlib.pyplot as plt

# Pontos dados
x0, x1, x2 = 1, 4, 6
f0 = np.log(x0)       # ln(1) = 0
f1 = np.log(x1)       # ln(4) = 1,386294
f2 = np.log(x2)       # ln(6) = 1,791759

#Diferenças divididas
f01 = (f1 - f0) / (x1 - x0)
f12 = (f2 - f1) / (x2 - x1)
f012 = (f12 - f01) / (x2 - x0)

# Função de interpolação de Newton de grau 2
def P(x):
    return f0 + f01*(x - x0) + f012*(x - x0)*(x - x1)

# Estimativa de ln(2)
x_estimativa = 2
ln2_est = P(x_estimativa)

print(f"Estimativa de ln(2) usando interpolação de Newton: {ln2_est:.6f}")

# Dados para gráfico
x = np.linspace(0, 6, 100)
y = P(x)

# Plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Interpolação de Newton (grau 2)', color='blue')
plt.plot(x_estimativa, ln2_est, 'ro', label=f'Estimativa ln(2) ≈ {ln2_est:.2f}')
plt.xlabel('x')
plt.ylabel('ln(x)')
plt.title('Estimativa de ln(2) com Interpolação de Newton')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()


```

Exemplo de relação entre Interpolação linear de Newton e Lagrange

```python
import numpy as np
import matplotlib.pyplot as plt

# Pontos
x1, x2 = 1, 4
f1 = 0
f2 = 1.386294

# Interpolação linear de Newton
def newton_linear(x):
    return f1 + (f2 - f1) / (x2 - x1) * (x - x1)

# Interpolação linear de Lagrange
def lagrange_linear(x):
    return f1 * (x - x2) / (x1 - x2) + f2 * (x - x1) / (x2 - x1)

# Estimativa em x = 2
x_est = 2
newtonval = newton_linear(x_est)
lagrange_val = lagrange_linear(x_est)

# Plot
x_vals = np.linspace(0, 5, 100)
plt.plot(x_vals, newton_linear(x_vals), 'b--', label='Newton Linear')
plt.plot(x_vals, lagrange_linear(x_vals), 'g:', label='Lagrange Linear')
plt.plot([x1, x2], [f1, f2], 'ko', label='Pontos dados')
plt.plot(x_est, newton_val, 'ro', label=f'Estimativa ln(2) ≈ {newton_val:.4f}')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('ln(x)')
plt.title('Interpolação Linear: Newton vs Lagrange')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Dados da questão 3
x = np.array([0.0, 1.8, 5.0, 6.0, 8.2, 9.2, 12.0])
y = np.array([26.000, 16.415, 5.375, 3.500, 2.015, 2.540, 8.000])

# Função de Lagrange
def lagrange_interpolacao(x_ponto, x_vals, y_vals):
    total = 0
    n = len(x_vals)
    for i in range(n):
        termo = y_vals[i]
        for j in range(n):
            if i != j:
                termo *= (x_ponto - x_vals[j]) / (x_vals[i] - x_vals[j])
        total += termo
    return total

# Interpolação em x = 3.5
x_desejado = 3.5
y_resultado = lagrange_interpolacao(x_desejado, x, y)

print(f"Valor interpolado em x = {x_desejado:.1f} é y = {y_resultado:.5f}")

# Geração de pontos para o gráfico de interpolação
x_plot = np.linspace(min(x), max(x), 300)
y_plot = [lagrange_interpolacao(xi, x, y) for xi in x_plot]

# Plotagem
plt.figure(figsize=(8, 5))
plt.plot(x_plot, y_plot, label='Interpolação de Lagrange', color='blue')
plt.scatter(x, y, color='red', label='Pontos fornecidos')
plt.scatter(x_desejado, y_resultado, color='green', label=f'f({x_desejado}) = {y_resultado:.5f}')
plt.title('Interpolação de Lagrange')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# Função de interpolação de Newton (com diferenças divididas)
def interp_newton(x_vals, y_vals, x_desejado):
    n = len(x_vals)
    coef = np.copy(y_vals)

    for j in range(1, n):
        coef[j:n] = (coef[j:n] - coef[j-1:n-1]) / (x_vals[j:n] - x_vals[0:n-j])

    p = coef[n-1]
    for k in range(n-2, -1, -1):
        p = p * (x_desejado - x_vals[k]) + coef[k]

    return p

# Dados da questão
x_full = np.array([1.0, 2.0, 2.5, 3.0, 4.0, 5.0])
y_full = np.array([0.0, 5.0, 6.5, 7.0, 3.0, 1.0])
x_interp = 3.4

x1 = np.array([3.0, 4.0])
y1 = np.array([7.0, 3.0])
x2 = np.array([2.5, 3.0, 4.0])
y2 = np.array([6.5, 7.0, 3.0])
x3 = np.array([2.5, 3.0, 4.0, 5.0])
y3 = np.array([6.5, 7.0, 3.0, 1.0])

# Avaliação
y1_interp = interp_newton(x1, y1, x_interp)
y2_interp = interp_newton(x2, y2, x_interp)
y3_interp = interp_newton(x3, y3, x_interp)

# Pontos para gráfico
x_plot = np.linspace(2.4, 4.6, 200)
y1_plot = [interp_newton(x1, y1, xi) for xi in x_plot]
y2_plot = [interp_newton(x2, y2, xi) for xi in x_plot]
y3_plot = [interp_newton(x3, y3, xi) for xi in x_plot]

# Função para formatar com vírgula
def formatar_virgula(x, pos):
    return f'{x:.1f}'.replace('.', ',')

# Plotagem
plt.figure(figsize=(9, 6))
plt.plot(x_plot, y1_plot, label='Grau 1', linestyle='--', color='orange')
plt.plot(x_plot, y2_plot, label='Grau 2', linestyle='-.', color='blue')
plt.plot(x_plot, y3_plot, label='Grau 3', linestyle='-', color='green')

plt.scatter(x_full, y_full, color='black', label='Pontos dados')
plt.scatter(x_interp, y1_interp, color='orange', s=50)
plt.scatter(x_interp, y2_interp, color='blue', s=50)
plt.scatter(x_interp, y3_interp, color='green', s=50)

# Aplicar formatador com vírgula
plt.gca().xaxis.set_major_formatter(FuncFormatter(formatar_virgula))
plt.gca().yaxis.set_major_formatter(FuncFormatter(formatar_virgula))

# Estética
plt.title('Interpolação de Newton')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
