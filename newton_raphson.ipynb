```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo a função e sua derivada
def f(x):
    return np.exp(-x) - x

def df(x):
    return -np.exp(-x) - 1

# Aproximação Inicial
x0 = 0.0

# Critério de Parada (6 dígitos significativos)
n = 6
Eppara = 0.5 * 10**(2-n)

# Pré-alocar informações
Epest = 100.0  # Erro percentual estimado inicial
x_old = 50

while Epest > Eppara :

    x_new = x_old - f(x_old)/df(x_old)

    # Calcula erro percentual
    if x_new != 0:  # Evita divisão por zero
        Epest = abs((x_new - x_old)/x_new) * 100

    # Exibe informações
    print(f"{x_old:.8f} | {f(x_old):.8f} | {Epest:.8f}")

    # Atualiza para próxima iteração
    x_old = x_new
```
