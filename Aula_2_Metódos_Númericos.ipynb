```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math

# Definindo o valor de X
x = 1

# valor verdadeiro
u = 2.71828182846

# definição da série de McLaurin
def serie_mclaurin(x, n):
    return (x**n) / math.factorial(n)

# definindo variáveis
N = 6
soma = 0
estimativa = []
EPT = []
nn = []

# laço for
for n in range(0, N):
    soma = soma + serie_mclaurin(x, n)
    ept = abs((u - soma) / soma) * 100
    estimativa.append(soma)
    EPT.append(ept)
    nn.append(n)

# criando o DataFrame
table = pd.DataFrame()
table['n'] = nn
table['estimativa'] = estimativa
table['ept'] = EPT




```

```python
# Gerando o gráfico
plt.plot(table['n'], table['estimativa'], 'o--', color='red')
plt.xlabel('Número de Interações')
plt.ylabel('Estimativa')
plt.grid(True)
plt.legend(['Estimativa da Série de McLaurin'])

# Exibindo o gráfico
plt.show()
```

```python
# Gerando o gráfico
plt.plot(table['n'], table['ept'], 'o--', color='blue', label='ept(%)')
plt.xlabel('Número de Interações')
plt.ylabel('$e_(pt)$ (%)')
plt.grid(True)
plt.legend(['Estimativa da Série de McLaurin'])

# Exibindo o gráfico
plt.show()
```
