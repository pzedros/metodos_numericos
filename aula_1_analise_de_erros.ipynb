```python
import numpy as np
import math
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Definindo o valor de X
x = 1
#Valor verdadeiro
u = 2.71828182846
#Definição da serie de Maclaurin
def serie(x,n):
  return (x**n)/math.factorial(n)

#Definindo Variáveis
N = 20
soma = 0
v_old = soma
estimativa = []
EPT = []
EPEST = []
nn = []

#Laço for
for n in range(0,N):
  soma = soma + serie(x,n)
  v_new = soma

  Ept = abs((u-soma)/u)*100 #erro percentual verdadeiro
  Epest = abs((v_new-v_old)/v_new)*100 #erro percentual estimado

  v_old = v_new
  estimativa.append(soma)
  EPT.append(Ept)
  EPEST.append(Epest)
  nn.append(n)


#Data frame
table = pd.DataFrame()
table['n'] = nn
table['Estimativa'] = estimativa
table['Ept'] = EPT
table['Epest'] = EPEST
print(table)
```

```python
#Grafico do Data Frame
plt.plot(table['n'], table['Estimativa'], 'o--', color='red', label='Estimativa')
plt.xlabel('Número de iterações')
plt.ylabel('Estimativa')
plt.grid()
plt.legend()
```

```python
#Grafico do Data Frame
plt.plot(table['n'], table['Ept'], 'o--', color='b', label='$E_{pt}$ (%)')
plt.plot(table['n'], table['Epest'], 'o--', color='r', label='$E_{pest}$ (%)')
plt.xlabel('Número de iterações')
plt.ylabel('$E_{pt}$ (%)')
plt.grid()
plt.legend()
```
