Questão 2 - exponencial

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0.75, 2.00, 3.00, 4.00, 6.00, 8.00, 8.50])
y = np.array([1.20, 1.95, 2.00, 2.40, 2.40, 2.70, 2.60])

Y = np.log(y)
n = len(x)
sum_xi = np.sum(x)
sum_xi_2 = np.sum(x**2)
sum_Y = np.sum(Y)
sum_xi_Yi = np.sum(x * Y)

A = np.array([[n, sum_xi], [sum_xi, sum_xi_2]])
b = np.array([[sum_Y], [sum_xi_Yi]])

Ai = np.linalg.inv(A) @ b

A0 = Ai[0][0]
beta1 = Ai[1][0]
alpha1 = np.exp(A0)

#calculo do R^2
y_medio = np.sum(y)/n
Sr = np.sum((y - alpha1 * np.exp(beta1 * x))**2)
St = np.sum((y - y_medio)**2)
Syx = np.sqrt(Sr/(n-2))

R_2 = (St - Sr)/St
print(f'R² = {R_2:.4f}')
print(f'Syx = {Syx:.4f}')
print(f'α₁ = {alpha1:.4f}, β₁ = {beta1:.4f}')

plt.plot(x, y, 'or', label = 'Dados Experimentais')
xx=np.linspace(min(x), max(x), 100)
plt.plot(xx, alpha1 * np.exp(beta1 * xx), '-b', label = '$y = \\alpha_1 e^{\\beta_1 x}$')
plt.legend()
plt.grid()
plt.ylabel('$y(x)$')
plt.xlabel('$x$')
plt.title('Ajuste Exponencial')
plt.show()
```

potencia simples

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0.75, 2.00, 3.00, 4.00, 6.00, 8.00, 8.50])
y = np.array([1.20, 1.95, 2.00, 2.40, 2.40, 2.70, 2.60])

Y = np.log(y)
X = np.log(x)
n = len(x)
sum_Xi = np.sum(X)
sum_Xi_2 = np.sum(X**2)
sum_Y = np.sum(Y)
sum_Xi_Yi = np.sum(X * Y)

A = np.array([[n, sum_Xi], [sum_Xi, sum_Xi_2]])
b = np.array([[sum_Y], [sum_Xi_Yi]])

ai = np.linalg.inv(A) @ b

A0 = ai[0][0]
beta2 = ai[1][0]
alpha2 = np.exp(A0)

#calculo do R^2
y_medio = np.sum(y)/n
Sr = np.sum((y - (alpha2 * x**beta2))**2)
St = np.sum((y - y_medio)**2)
Syx = np.sqrt(Sr/(n-2))

R_2 = (St - Sr)/St
print(f'R² = {R_2:.4f}')
print(f'Syx = {Syx:.4f}')
print(f'α₁ = {alpha2:.4f}, β₁ = {beta2:.4f}')

plt.plot(x, y, 'or', label = 'Dados Experimentais')
xx=np.linspace(min(x), max(x), 100)
plt.plot(xx, alpha2 * xx**beta2, '-g', label = '$y = \\alpha_2 x^{\\beta_2}$')
plt.legend()
plt.grid()
plt.ylabel('$y(x)$')
plt.xlabel('$x$')
plt.title('Ajuste de Potência Simples')
plt.show()
```

parábola

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0.75, 2.00, 3.00, 4.00, 6.00, 8.00, 8.50])
y = np.array([1.20, 1.95, 2.00, 2.40, 2.40, 2.70, 2.60])

n = len(x)
sum_xi = np.sum(x)
sum_yi = np.sum(y)
sum_xi_2 = np.sum(x**2)
sum_xi_3 = np.sum(x**3)
sum_xi_4 = np.sum(x**4)
sum_xi_yi = np.sum(x*y)
sum_xi2_yi = np.sum((x**2)*y)

A = np.array([[n, sum_xi, sum_xi_2], [sum_xi, sum_xi_2, sum_xi_3], [sum_xi_2, sum_xi_3, sum_xi_4]])
b = np.array([[sum_yi], [sum_xi_yi], [sum_xi2_yi]])

ai = (np.linalg.inv(A) @ b).flatten()

#calculo do R^2
y_medio = np.sum(y)/n
Sr = np.sum((y - ai[0] - ai[1]*x - ai[2]*x**2)**2)
St = np.sum((y - y_medio)**2)
Syx = np.sqrt(Sr/(n-3))

R_2 = (St - Sr)/St
print(f'R² = {R_2:.4f}')
print(f'Syx = {Syx:.4f}')
print(f'a0 = {ai[0]:.4f}')
print(f'a1 = {ai[1]:.4f}')
print(f'a2 = {ai[2]:.4f}')


plt.plot(x, y, 'or', label = 'Dados Experimentais')
xx=np.linspace(min(x), max(x), 100)
plt.plot(xx, ai[0] + ai[1]*xx + ai[2]*xx**2, '-m', label = '$y=a_0+a_1x+a_2x^2$')
plt.legend()
plt.grid()
plt.ylabel('$y(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
import matplotlib.ticker as ticker
plt.plot(x, y, 'or', label = '$Dados\ Experimentais$')
xx=np.linspace(min(x), max(x), 100)
plt.plot(xx, alpha1 * np.exp(beta1 * xx), '-b', label = '$Exponencial$')
plt.plot(xx, alpha2 * xx**beta2, '-g', label = '$Potência\ Simples$')
plt.plot(xx, ai[0] + ai[1]*xx + ai[2]*xx**2, '-m', label = '$Parabólica$')

formatter = ticker.FuncFormatter(lambda val, pos: f'{val:.2f}'.replace('.', ','))

plt.gca().xaxis.set_major_formatter(formatter)
plt.gca().yaxis.set_major_formatter(formatter)

plt.legend()
plt.title('Ajustes de Curvas')
plt.grid()
plt.ylabel('$y(x)$')
plt.xlabel('$x$')
plt.show()
```
