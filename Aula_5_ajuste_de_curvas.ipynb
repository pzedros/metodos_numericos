Ajuste Linear

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 2, 4, 6, 9, 11, 12, 15, 17, 19])
y = np.array([5, 6, 7, 8, 9, 10, 10, 11, 12, 12])

n = len(x)
sum_xi = np.sum(x)
sum_xi_2 = np.sum(x**2)
sum_xi_yi = np.sum(x*y)
sum_yi = np.sum(y)

A = np.array([[n, sum_xi], [sum_xi, sum_xi_2]])
b = np.array([[sum_yi], [sum_xi_yi]])

ai = np.linalg.inv(A) @ b

#calculo do R^2
y_medio = np.sum(y)/n
Sr = np.sum((y - ai[0] - ai[1]*x)**2)
St = np.sum((y - y_medio)**2)

R_2 = (St - Sr)/St
print(R_2)


plt.plot(x, y, 'or', label = 'Dados Experimentais')
plt.plot(x, ai[0] + ai[1]*x, '-b', label = '$y=a_0+a_1x+a_2x^2$')
plt.legend()
plt.grid()
plt.ylabel('$y(x)$')
plt.xlabel('$x$')
plt.show()
```

Regressão Linear Polinomial grau 2

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([2.1, 7.7, 13.6, 27.2, 40.9, 61.1])

n = len(x)
sum_xi = np.sum(x)
sum_yi = np.sum(y)
sum_xi_2 = np.sum(x**2)
sum_xi_3 = np.sum(x**3)
sum_xi_4 = np.sum(x**4)
sum_xi_yi = np.sum(x*y)
sum_xi2_yi = np.sum((x**2)*y)

A = np.array([[n, sum_xi, sum_xi_2], [sum_xi, sum_xi_2, sum_xi_3], [sum_xi_2, sum_xi_3, sum_xi_4]])
b = np.array([[sum_yi], [sum_xi_yi], [sum_xi2_yi]])

ai = np.linalg.inv(A) @ b

#calculo do R^2
y_medio = np.sum(y)/n
Sr = np.sum((y - ai[0] - ai[1]*x - ai[2]*x**2)**2)
St = np.sum((y - y_medio)**2)

R_2 = (St - Sr)/St
print(R_2)


plt.plot(x, y, 'or', label = 'Dados Experimentais')
plt.plot(x, ai[0] + ai[1]*x + ai[2]*x**2, '-b', label = '$y=a_0+a_1x$')
plt.legend()
plt.grid()
plt.ylabel('$y(x)$')
plt.xlabel('$x$')
plt.show()
```

Regrassão Linear Múltipla

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([460, 960, 2485, 1600, 1245])
x2 = np.array([200, 150, 50, 20, 10])
y = np.array([280, 320, 450, 500, 550])

n = len(x1)
sum_x1i = np.sum(x1)
sum_x2i = np.sum(x2)
sum_yi = np.sum(y)

sum_x1i_2 = np.sum(x1**2)
sum_x1i_x2i = np.sum(x1*x2)
sum_x1i_yi = np.sum(x1*y)

sum_x2i_2 = np.sum(x2**2)
sum_x2i_yi = np.sum(x2*y)

A = np.array([[n, sum_x1i, sum_x2i], [sum_x1i, sum_x1i_2, sum_x1i_x2i], [sum_x2i, sum_x1i_x2i, sum_x2i_2]])
b = np.array([[sum_yi], [sum_x1i_yi], [sum_x2i_yi]])

ai = np.linalg.inv(A) @ b

#calculo do R^2
y_medio = np.sum(y)/n
Sr = np.sum((y - ai[0] - ai[1]*x1 - ai[2]*x2)**2)
St = np.sum((y - y_medio)**2)

R_2 = (St - Sr)/St
print(R_2)

print(ai)

#Gráfico 3d
x1 = np.linspace(min(x1), max(x1), 100)
x2 = np.linspace(min(x2), max(x2), 100)

X1, X2 = np.meshgrid(x1, x2)
y = ai[0] + ai[1]*X1 + ai[2]*X2

plt.figure()
ax = plt.axes(projection= '3d')
cs = ax.plot_surface(X1, X2, y, cmap = 'viridis')
ax.set_xlabel('$x1$')
ax.set_ylabel('$x2$')
ax.set_zlabel('$y$')
ax.view_init(elev=30, azim=135)
plt.show()
```

Interpolação Linear de Newton

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 4])
f = np.array([0, 1.386294])
X = 2

f1 = f[0] + (((f[1]-f[0])/(x[1]-x[0])))*(X - x[0])
print(f1)

xa = np.linspace(0, 4, 100)

plt.plot(xa, np.log(xa))
plt.plot(X, f1, 'or')
```

Interpolação Quadrática de Newton

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 4, 6])
f = np.array([0, 1.386294, 1.791759])
X = 2

f2 = f[0] + ((f[1]-f[0])/(x[1]-x[0]))*(X-x[0]) + ((((f[2]-f[1])/(x[2]-x[1])) - ((f[1]-f[0])/(x[1]-x[0])))/(x[2]-x[0]))*(X - x[0])*(X - x[1])
print(f2)

xa = np.linspace(0, 6, 100)

plt.plot(xa, np.log(xa))
plt.plot(X, f2, 'or')
```
