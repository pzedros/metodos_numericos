B) uma unica aplicacao da regra do Trapezio

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

def f(x):
  return 1 + x + 2*x**3 - x**4 + 5*x**5

a = -1/2
b = 1
#valor verdadeiro
u = 2.9578


IT = (b-a)*(f(a)+f(b))/2
print(f'integral: {IT}')
Ept_IT = abs((u - IT)/u) * 100
print(f'erro: {Ept_IT}')

x = np.linspace(a,b,100)
plt.plot(x, f(x), '-b')
plt.grid()
plt.fill_between(x, f(x), alpha=0.3, color='skyblue', label='Área sob a curva')

formatter = ticker.FuncFormatter(lambda val, pos: f'{val:.2f}'.replace('.', ','))

plt.gca().xaxis.set_major_formatter(formatter)


plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da Integral')
plt.legend()
plt.show()
```

C) multiplas da regra do Trapezio, n=2

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return 1 + x + 2*x**3 - x**4 + 5*x**5

x0 = -1/2
xn = 1
n = 2 #numero de trapezios
h = (xn - x0)/n

#valor verdadeiro
u = 2.9578
soma = 0

for i in range(1, n):
  xi = x0 + h*i
  soma = soma + f(xi)

  ITM = h/2 * (f(x0) + 2*soma + f(xn))
  Ept_ITM = abs((u - ITM)/u) * 100

print(ITM)
print(Ept_ITM)
```

**E**) multipla da regra 1/3 de Simpson, n=2

```python
import numpy as np

# Valor exato da integral para comparação
exact_value = 2.9578125

def f(x):
    return 1 + x + 2*x**3 - x**4 + 5*x**5

def simpson_1_3_multiple(func, a, b, n):
    if n % 2 != 0:
        raise ValueError("n deve ser par para a Regra 1/3 de Simpson.")

    h = (b - a) / n
    soma = func(a) + func(b)

    for i in range(1, n):
        x = a + i * h
        if i % 2 == 0:
            soma += 2 * func(x)
        else:
            soma += 4 * func(x)

    return (h / 3) * soma

# Limites de integração
a = -0.5
b = 1

# Aplicação com n = 2
n = 2
I_n2 = simpson_1_3_multiple(f, a, b, n)
erro_n2 = abs((exact_value - I_n2) / exact_value) * 100

print(f"Integral (Simpson 1/3, n=2): {I_n2}")
print(f"Erro percentual: {erro_n2:.8f}%")
```
