```python
import numpy as np
import math
import matplotlib.pyplot as plt
import pandas as pd

#definindo x
x = 1
#definindo a função
def serie(x, i):
  return (x**i)/math.factorial(i)

#valor verdadeiro
u = 2.71828182846

#numero de algaritimos significativos
n = 6

#criterio de Scarvought(1966)
Eppara = 0.5*10**(2-n)

#definindo variaveis
soma = 0
v_old = soma
i = 0
Ept = 100
Epest = 100
estimativa = []
EPT = []
EPEST = []
nn = [] #numero de iteraçoes

while Epest >= Eppara:
  soma = soma + serie(x,i)   #somar a serie
  v_new = soma
  Ept = abs((u - soma)/u)*100
  Epest = abs((v_new - v_old)/v_new)*100
  i = i + 1
  v_old = v_new

  #salvando os calculos
  nn.append(i)
  estimativa.append(soma)
  EPT.append(Ept)
  EPEST.append(Epest)

#data frame
table = pd.DataFrame()
table['n'] = nn
table['Estimativa'] = estimativa
table['Ept'] = EPT
table['Epest'] = EPEST
print(table)
```
